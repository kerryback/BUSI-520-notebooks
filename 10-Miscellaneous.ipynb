```python
# Passing by value

x = 2
y = x
y = 10
print(f"x = {x}")
print(f"y = {y}")
```

```python
# Passing by reference

x = [2, 2]
y = x
y[0] = 10
print(f"x = {x}")
print(f"y = {y}")
```

```python
# Make a copy to force passing by value

x = [2, 2]
y = x.copy()
y[0] = 10
print(f"x = {x}")
print(f"y = {y}")
```

```python
# Same issue when altering within a function

def func(arg):
    arg[0] = 10
    return arg

x = [2, 2]
y = func(x)
print(f"x = {x}")
print(f"y = {y}")
```

```python
# Again, make a copy

def func(arg):
    new_var = arg.copy()
    new_var[0] = 10
    return new_var

x = [2, 2]
y = func(x)
print(f"x = {x}")
print(f"y = {y}")
```

```python
# Local variable

def func(arg):
    new_var = arg.copy()
    new_var[0] = 10
    return new_var

x = [2, 2]
y = func(x)
print(f"new_var = {new_var}")
```

```python
# Local variable again

x = 2

def func(arg):
    x = 10
    return x*arg

y = func(3)

print(f"x = {x}")
print(f"y = {y}")
```

```python
# Global variable

x = 2

def func(arg):
    return x*arg

y = func(3)

print(f"x = {x}")
print(f"y = {y}")
```

```python
# Local variable one more time

x = [2, 2]

def func(arg):
    x = 2*x
    return x*arg

y = func(3)

print(f"x = {x}")
print(f"y = {y}")
```

### Summary

Be careful of two things:

1.  If you pass a list, dictionary, numpy array, pandas series, or pandas dataframe to a function and the function argument is modified within the function, then it will also be modified outside the function.  So, make a copy if you don't want the outside version to be modified.

2.  If you intend to use a local variable inside a function but forget to define it, and a variable with the same name exists outside the function, then the outside variable will be used in the function.
